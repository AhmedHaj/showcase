Write function to resize images:

```python
# takes in an array of numpy arrays, returns an array with resized numpy array representations
def resize_sample(original_array, rows, cols):
    resized_array = []
    
    for i in original_array:
        resized_array.append(transform.resize(i, (rows, cols)))
    
    return np.asarray(resized_array)
```

```python
# takes in an array of numpy arrays, returns an array with grayscaled numpy array representations
def gray_sample(array):
    gray_array = []
    
    for i in array:
        gray_array.append(color.rgb2gray(i))
    
    return np.asarray(gray_array)
```

Concatenating sample data to create the datasets for the m/n classifier and the s/5 classifier:

Since we will be using a logistic regression classifier, it is extremely important that we use balanced datasets. Otherwise the results will be skewed in favor of the majority class because the majority class will have a higher probability of occuring. 

To combat class imbalance, I will be employing a naive implementation of undersampling where I discard the extra samples of the majority classes to have a balanced dataset for both binary classifiers.


```python
x1 = np.concatenate((m_samples_gray, n_samples_gray[:43])) # m/n dataset while implementing naive undersampling
x2 = np.concatenate((s_samples_gray[:64], five_samples_gray)) # s/5 dataset while implementing naive undersampling

y1 = np.concatenate((m_labels, n_labels[:43]))
y2 = np.concatenate((s_labels[:64], five_labels))
```

## Test and Training Data

### m/n Classifier:

Since we have taken care of the class imbalance in our dataset for both classifiers, we should not need to worry too much about stratifying our datasets. However, due to the small sample size of our dataset, it would be beneficial to employ stratified sampling to ensure that we keep the 50/50 class balance so that our model can learn both classes equally.

In addition, I will be using Stratified K fold (5 folds) to perform cross-validation while ensuring that there are m and n samples in both the training and the validation sets. 


Our dataset will be organized into 60% training, 20% validation, and 20% testing.

For the m/n classifier: 50 training, 18 validation, and 18 testing.

For the s/5 classifier: 76 training, 26 validation, and 26 testing.


```python
from sklearn.model_selection import train_test_split

# m/n Classifier 
# stratify our dataset to ensure all sets are balanced
X_train, X_test, y_train, y_test = train_test_split(x1, y1, stratify=y1, test_size=0.2, random_state=42)

print("The m/n classifier dataset has {0} training samples (before training-validation split) and {1} testing samples.".format(y_train.shape[0], y_test.shape[0]))


# Reshaping the training and testing sample into 2d shapes to be able to work on our model
nsamples, nx, ny = X_train.shape
X_train = X_train.reshape((nsamples, -1))

nsamples, nx, ny = X_test.shape
X_test = X_test.reshape((nsamples, -1))

# s/5 Classifier 
# stratify our dataset to ensure all sets are balanced
X_train2, X_test2, y_train2, y_test2 = train_test_split(x2, y2, stratify=y2, test_size=0.2, random_state=42)

print("The s/5 classifier dataset has {0} training samples (before training-validation split) and {1} testing samples.".format(y_train2.shape[0], y_test2.shape[0]))


# Reshaping the training and testing sample into 2d shapes to be able to work on our model
nsamples, nx, ny = X_train2.shape
X_train2 = X_train2.reshape((nsamples,-1))


nsamples, nx, ny = X_test2.shape
X_test2 = X_test2.reshape((nsamples,-1))
```

Creating our Logistic Regression Models, and training them:

```python
from sklearn.linear_model import LogisticRegression
#from sklearn.base import clone

# m/n classifier
bin_clf = LogisticRegression(max_iter=1000, tol=0.1, solver='liblinear')
bin_clf.fit(X_train, y_train)


# s/5 classifier
bin_clf2 = LogisticRegression(max_iter=1000, tol=0.1, solver='liblinear')
bin_clf2.fit(X_train2, y_train2)
```

Cross-validating our model on 5 folds using cross_val_score. (after testing cross_val_score and StratifiedKFold, I found out that they output the exact same values as far the context of the assignment is concerned. I will be using cross_val_score for easier readability)



```python
from sklearn.model_selection import cross_val_score
#from sklearn.model_selection import StratifiedKFold

# m/n classifier
cv_score = cross_val_score(bin_clf, X_train, y_train, cv=5, scoring="accuracy")

print(cv_score)
print("Average accuracy of m/n classifier during cross-validation = {0}".format(sum(cv_score)/5))

print()
# s/5 classifier
cv_score2 = cross_val_score(bin_clf2, X_train2, y_train2, cv=5, scoring="accuracy")
print(cv_score2) 
print("Average accuracy of s/5 classifier during cross-validation = {0}".format(sum(cv_score2)/5))




#skf = StratifiedKFold(n_splits = 5)
#for train, test in skf.split(X_train, y_train):
#    clone_clf = clone(bin_clf)
#    X_train_fold = X_train[train]
#    y_train_fold = y_train[train]
#    
#    X_test_fold = X_train[test]
#    y_test_fold = y_train[test]
    
#    clone_clf.fit(X_train_fold, y_train_fold)
#    y_pred = clone_clf.predict(X_test_fold)
#    n_correct = sum(y_pred == y_test_fold)
#    print(n_correct / len(y_pred))
```

## Plotting the ROC Curve



```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_curve


# m/n classifier ROC curve
y_scores = cross_val_predict(bin_clf, X_train, y_train, cv=5, method="decision_function")

fpr, tpr, thresholds = roc_curve(y_train, y_scores)

def plot_roc_curve(fpr, tpr, title, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
    plt.title(title)

plt.figure(figsize=(8, 6))
plot_roc_curve(fpr, tpr, "m/n Classifier")
plt.show()

# s/5 classifier ROC Curve 
y_scores = cross_val_predict(bin_clf2, X_train2, y_train2, cv=5, method="decision_function")


fpr, tpr, thresholds = roc_curve(y_train2, y_scores, pos_label = 5)
plt.figure(figsize=(8, 6))
plot_roc_curve(fpr, tpr, "s/5 Classifier")
plt.show()


```

# 1.3 Multiclass Classifier

## Organizing the data:

The datasets used in training each of the binary classifiers will be concatenated to create the dataset that the Multiclass Classifier will use.

The naive undersampling approach will be used to ensure that our dataset is balanced.

```python
# create dataset while implementing naive undersampling
multi_dataset = np.concatenate((m_samples_gray, n_samples_gray[:43], s_samples_gray[:43], five_samples_gray[:43])) 
y3 = np.concatenate((m_labels, n_labels[:43], s_labels[:43], five_labels[:43]))

# plot 4 images to show that the data was concatenated successfully:

indexM = np.where(y3==0)[0]
indexN = np.where(y3==1)[0]
indexS = np.where(y3==4)[0]
index5 = np.where(y3==5)[0]


pickM = indexM[np.random.choice(indexM.shape[0])]
pickN = indexN[np.random.choice(indexN.shape[0])]
pickS = indexS[np.random.choice(indexS.shape[0])]
pick5 = index5[np.random.choice(index5.shape[0])]

figure, axes = plt.subplots(nrows=4, ncols=1)
ax = axes.ravel()


ax[0].imshow(multi_dataset[pickM], cmap='gray')
ax[0].set_title("m image")

ax[1].imshow(multi_dataset[pickN] , cmap='gray')
ax[1].set_title("n image")

ax[2].imshow(multi_dataset[pickS] , cmap='gray')
ax[2].set_title("s image")

ax[3].imshow(multi_dataset[pick5] , cmap='gray')
ax[3].set_title("5 image")
```

##### Test and Training data

Similar to before, the data will have a 60-20-20 split. That is, we will have 102 training, 35 validation, and 35 testing. 

In addition, I will be using Stratified K fold (5 folds) to perform cross-validation, ensuring that each of the datasets contains a balanced percentage of classes.

```python
# stratify our dataset to ensure the dataset stays balanced
X_train, X_test, y_train, y_test = train_test_split(multi_dataset, y3, stratify=y3, test_size=0.2, random_state=42)

print("The multiclass classifier dataset has {0} training samples (before training-validation split) and {1} testing samples.".format(y_train.shape[0], y_test.shape[0]))

nsamples, nx, ny = X_train.shape
X_train = X_train.reshape((nsamples,-1))

nsamples, nx, ny = X_test.shape
X_test = X_test.reshape((nsamples, -1))
```

Creating and training our model:

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train)

cv_score = cross_val_score(sgd_clf, X_train, y_train, cv=5, scoring="accuracy")

print(cv_score)
print("Average accuracy of multiclass classifier during cross-validation = {0}".format(sum(cv_score)/5))
```

A low average accuracy tells us that our model is not performing so well. However, it is greater than 25% accuracy that we would get when randomly guessing since we have 4 classes. However further info is required to see what the model is doing incorrectly, so we should print the confusion matrix:



```python
# testing against training set 
y_pred = sgd_clf.predict(X_train)

print('Multiclass Classifier scores on training set: ')
print("Accuracy is {0}".format(accuracy_score(y_train, y_pred)))
print( 'And the confusion matrix: \n {0}'.format(confusion_matrix(y_train, y_pred) ))

print()

# testing against testing set
y_pred = sgd_clf.predict(X_test)

print('Multiclass Classifier scores on testing set: ')
print("Accuracy is {0}".format(accuracy_score(y_test, y_pred)))
print( 'And the confusion matrix: \n {0}'.format(confusion_matrix(y_test, y_pred) ))
```

## Part II: Using edge detection on images to highlight shapes of letters/number

Since we are attempting to classify shapes, I believe that preprocessing the images in such a way that the edges of the shapes are highlighted, and everything else is discarded, would make our models perform better because they will have random noise in the images to be confused by.

By removing random noise such as the color inside the shape (present in grayscale images too), and the color outside the shape, our model should perform better.

```python
# We know that the first 86 samples are from the m/n samples
edge_binary_dataset = edge_multi_dataset[:86]
y1 = y3[:86]

# stratify our dataset to ensure that all sets are balanced
X_train, X_test, y_train, y_test = train_test_split(edge_binary_dataset, y1, stratify=y1, test_size=0.2, random_state=42)

print("The m/n classifier dataset has {0} training samples (before training-validation split) and {1} testing samples.".format(y_train.shape[0], y_test.shape[0]))


# Reshaping the training and testing sample into 2d shapes to be able to work on our model
nsamples, nx, ny = X_train.shape
X_train = X_train.reshape((nsamples, -1))

nsamples, nx, ny = X_test.shape
X_test = X_test.reshape((nsamples, -1))



# m/n classifier
bin_clf = LogisticRegression(max_iter=1000, tol=0.1, solver='liblinear')
bin_clf.fit(X_train, y_train)



# testing against training set 
y_pred = bin_clf.predict(X_train)

print('m/n Classifier scores on training set: ')
print("Accuracy is {0}".format(accuracy_score(y_train, y_pred)))
print('Precision is {0} '.format(precision_score(y_train, y_pred)))
print('Recall is {0} '.format(recall_score(y_train, y_pred)))
print( 'And the confusion matrix: \n {0}'.format(confusion_matrix(y_train, y_pred) ))

print()

# testing against testing set
y_pred = bin_clf.predict(X_test)

print('m/n Classifier scores on testing set: ')
print("Accuracy is {0}".format(accuracy_score(y_test, y_pred)))
print('Precision is {0} '.format(precision_score(y_test, y_pred)))
print('Recall is {0} '.format(recall_score(y_test, y_pred)))
print( 'And the confusion matrix: \n {0}'.format(confusion_matrix(y_test, y_pred) ))


# m/n classifier ROC curve
y_scores = cross_val_predict(bin_clf, X_train, y_train, cv=5, method="decision_function")

fpr, tpr, thresholds = roc_curve(y_train, y_scores)

def plot_roc_curve(fpr, tpr, title, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
    plt.title(title)

plt.figure(figsize=(8, 6))
plot_roc_curve(fpr, tpr, "m/n Classifier")
plt.show()

```

This could be for a number of reasons, possibly the n samples chosen in the training set being more similar and the m samples being more diverse. However, at 88% accuracy, this model is suitable for performing the task of differentiating between m and n samples.
